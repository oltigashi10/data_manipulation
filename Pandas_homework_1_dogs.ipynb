```python
import difflib
import pandas as pd
import csv
import argparse

parser = argparse.ArgumentParser(description='this is the initial parser')
parser.add_argument('--input', type=str, required=True)
parser.add_argument('--output', type=str, required=True)
args = parser.parse_args()
assigned_disciplines = []
def read_and_compare_csv(in_file: str, out_file: str):
    df = pd.read_csv(in_file)
    programs_column, disciplines_column = df.programname.to_list(), df.discipline.to_list()
    for i, j in enumerate(programs_column):
        new_d = [(k, round(difflib.SequenceMatcher(None, j, k).ratio()*100, 1)) for k in disciplines_column]
        assigned_disciplines.append([', '.join(map(str, x)) for x in sorted(new_d, key=lambda x:x[1], reverse=True)[:5]])
    data = assigned_disciplines
    with open(out_file,'w') as out:
        csv_out=csv.writer(out)
        csv_out.writerow(['suggested_one','suggested_2', 'suggested_3', 'suggested_4', 'suggested_5'])
        for row in data:
            csv_out.writerow(row)

if __name__=='__main__':
    read_and_compare_csv(args.input, args.output)
```
